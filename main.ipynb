```python
x = data[['Car Purchase Amount']]
x
```

```python
y= data['Annual Salary']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train , y_test = train_test_split(x, y, test_size=0.2, random_state=True)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept=True)
```

```python
regressor.fit(x_train, y_train)
```

```python
print('Linear Coefficient (m)', regressor.coef_)
print('Linear Coefficient (b)', regressor.intercept_)
```

```python
x_test
```

```python
y_predicted = regressor.predict(x_test)
```

```python
y_predicted
```

```python
y_test
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, regressor.predict(x_train), color = 'green')
plt.xlabel('Purchase Amount')
plt.ylabel('Annual Salary')
plt.title('The relationship between Car Purchase Amount and a persons salary')
```
